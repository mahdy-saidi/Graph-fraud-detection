# Nouvelle base de données de simulations de transactions 

```python
print('='*60)
print('          Train Fraud')
print('='*60)
train_fraud = pd.read_csv("../data/SimulatedfraudTrain.csv")
print(f"Nombre de transactions: {len(train_fraud):,}")
print(f"Nombre de variables: {train_fraud.shape[1]}")
print(train_fraud.describe())

# valeurs manquantes
print ('     valeurs manquantes ')
missing = train_fraud.isnull().sum()
missing_pct = (missing / len(train_fraud)) * 100
missing_df = pd.DataFrame({
    'Colonne': missing.index,
    'Valeurs_Manquantes': missing.values,
    'Pourcentage': missing_pct.values
})
missing_df = missing_df[missing_df['Valeurs_Manquantes'] > 0].sort_values('Valeurs_Manquantes', ascending=False)
if len(missing_df) > 0:
    print("Colonnes avec valeurs manquantes:")
    print(missing_df.to_string(index=False))
    
    # Visualisation
    if len(missing_df) > 0:
        plt.figure(figsize=(12, 6))
        plt.barh(missing_df['Colonne'], missing_df['Pourcentage'], color='coral')
        plt.xlabel('Pourcentage de valeurs manquantes')
        plt.title('Valeurs Manquantes par Variable', fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()
else:
    print("Aucune valeur manquante détectée!")


# Fraud 

# Compter les valeurs 0 et 1
print("Nombre de transactions par type:")
print(train_fraud['is_fraud'].value_counts())

print("\nPourcentages:")
print(train_fraud['is_fraud'].value_counts(normalize=True) * 100)

# Ou de façon plus détaillée
nb_non_fraude = (train_fraud['is_fraud'] == 0).sum()
nb_fraude = (train_fraud['is_fraud'] == 1).sum()

print(f"\n0 (Non-Fraude): {nb_non_fraude:,}")
print(f"1 (Fraude): {nb_fraude:,}")
print(f"Total: {len(train_fraud):,}")

# gender 

print("Nombre de transactions pour chaque type de genre")
print(train_fraud['gender'].value_counts())

print("\nPourcentages:")
print(train_fraud['gender'].value_counts(normalize=True) * 100)

# Ou de façon plus détaillée
nb_masculin = (train_fraud['gender'] == 'M').sum()
nb_feminin = (train_fraud['gender'] == 'F').sum()

print(f"\nM (Masculin): {nb_masculin:,}")
print(f"F (Féminin): {nb_feminin:,}")
print(f"Total: {len(train_fraud):,}")


# Vérification de l'uniicité de la colonne "cc_num"

# Nombre total de lignes
n_total = len(train_fraud)

# Nombre de valeurs uniques de cc_num
n_unique = train_fraud['cc_num'].nunique()

print("Nombre total de transactions :", n_total)
print("Nombre de cc_num uniques     :", n_unique)

if n_unique == n_total:
    print("La colonne cc_num est unique (une valeur par transaction).")
else:
    print("La colonne cc_num n'est PAS unique (une carte a plusieurs transactions).")


# vérification de l'unicité de la colonne 'trans_num'
# Nombre de valeurs uniques de trans_num
n_unique = train_fraud['trans_num'].nunique()

print("Nombre total de transactions :", n_total)
print("Nombre de trans_num uniques     :", n_unique)

if n_unique == n_total:
    print("La colonne trans_num est unique (une valeur par transaction).")
else:
    print("La colonne trans_num n'est PAS unique .")
```
